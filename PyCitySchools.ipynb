```python
import pandas as pd
```

```python
#to declare variables to import the datasets
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)
```

```python
school_data_df.count()
student_data_df.count()
```

```python
school_data_df.school_name.dtype
```

```python
#add the prefixes and suffixes to be removed to a new list
prefixes_suffixes = ["Dr. ", "Mr. ", "Mrs. ", "Ms. ","Miss ", " MD", " DDS"," DVM", " PhD" ]
```

```python
for names in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(names, "")
```

```python
student_data_df.head(10)
```

```python
#to check if data is successfully cleaned
prefixes = []
suffixes = []
```

```python
students_name = student_data_df["student_name"].tolist()
```

```python
for names in students_name:
    if len(names.split()[0]) <= 4:
        prefixes.append(names)
    if len(names.split()[-1]) <= 3:
        suffixes.append(names)
```

```python
set(prefixes)
set(suffixes)
```
